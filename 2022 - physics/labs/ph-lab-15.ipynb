```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array(range(-8, 9))
print(x)
```

```python
y_50 = np.array([0, 0, 0, 0, 0, 1, 5, 21, 9, 6, 6, 1, 0, 1, 0, 0, 0])
f_50 = y_50 / 50
h_50 = []
for n, i in enumerate(x):
    for e in range(y_50[n]):
        h_50.append(i)

print(y_50)
print(f_50)
print(h_50)



# mu, sigma = 115, 15
# x = mu + sigma * np.random.randn(10000)

fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')

# the histogram of the data
n, bins, patches = ax.hist(h_50, x, density=True, facecolor='C0', alpha=1.)

ax.set_xlabel('Ячейки')
ax.set_ylabel('Частота попадания')
ax.set_title('Распределение при N=50')
ax.axis([-8, 8, 0, 0.5])
ax.grid(True);
```

```python
y_100 = np.array([0, 0, 0, 0, 0, 5, 11, 25, 11, 20, 14, 7, 6, 1, 0, 0, 0])
f_100 = y_100 / 100
h_100 = []
for n, i in enumerate(x):
    for e in range(y_100[n]):
        h_100.append(i)

print(y_100)
print(f_100)
print(h_100)



# mu, sigma = 115, 15
# x = mu + sigma * np.random.randn(10000)

fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')

# the histogram of the data
n, bins, patches = ax.hist(h_100, x, density=True, facecolor='C0', alpha=1.)

ax.set_xlabel('Ячейки')
ax.set_ylabel('Частота попадания')
ax.set_title('Распределение при N=100')
ax.axis([-8, 8, 0, 0.5])
ax.grid(True);
```

```python
y_300 = np.array([0, 0, 0, 2, 7, 21, 30, 57, 44, 52, 52, 22, 5, 5, 2, 0, 1])
f_300 = y_300 / 300
h_300 = []
for n, i in enumerate(x):
    for e in range(y_300[n]):
        h_300.append(i)

h_300 = np.array(h_300)
        
print(y_300)
print(f_300)
print(h_300)

# mu, sigma = 115, 15
# x = mu + sigma * np.random.randn(10000)

fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')

# the histogram of the data
n, bins, patches = ax.hist(h_300, x, density=True, facecolor='C0', alpha=1.)

sigma = 2.2

mu = h_300.mean()

print(mu)


bins = np.arange(-8.0, 8.0, 0.1)

y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')

y = (0.19 / np.sqrt(np.e) + bins * 0)
ax.plot(bins, y, '--')

ax.axvline(x=-1.9, ymin=0, ymax=0.3, color='red')


ax.set_xlabel('Ячейки')
ax.set_ylabel('Частота попадания')
ax.set_title('Распределение при N=300')
ax.axis([-8, 8, 0, 0.5])
ax.grid(True);
```
