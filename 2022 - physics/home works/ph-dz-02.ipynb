## Задание 6

**В одной координатной системе** постройте графики функции Гауса и нормального распределния для $\sigma = 0,1$

```python
sigma = 0.1
fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')
x = np.arange(-0.4, 0.4, 0.001)

display(Markdown(rf'Распределение Гаусса: $F(x) = e ^ {{- \frac{{x^2}}{{2 \sigma }} }}$'))
display(Markdown(rf'Нормальное распределение: '
                 rf'$F(x) =  \frac {{1}}{{\sigma \sqrt {{ 2 \pi  }} }} e ^ {{- \frac{{x^2}}{{2 \sigma ^2 }} }}$'))

# Строим распределение Гаусса
y = (np.exp(- x**2 / (2 * sigma**2)))
ax.plot(x, y, label='Распределение Гаусса')

# Строим нормальное распределение
y = (np.exp(- x**2 / (2 * sigma**2))) / (sigma * (2 * np.pi)**0.5)
ax.plot(x, y, linestyle='--', label='нормальное распределение')

# Рисуем оси
ax.spines[['top', 'right']].set_visible(False)
ax.spines[['top', 'right']].set_visible(False)
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

# Рисуем стрелочки
arrow_fmt = dict(markersize=4, color='black', clip_on=False)
ax.plot((1), (0), marker='>', transform=ax.get_yaxis_transform(), **arrow_fmt)
ax.plot((0), (1), marker='^', transform=ax.get_xaxis_transform(), **arrow_fmt)

# Create 'x' and 'y' labels placed at the end of the axes
ax.set_xlabel('x', size=14, labelpad=0, x=1.)
ax.set_ylabel('F(x)', size=14, labelpad=0, y=1., rotation=0)

# Create custom major ticks to determine position of tick labels
x_frequency, y_frequency = 0.1, 1. 
xmin, xmax, ymin, ymax = -0.4, 0.4+x_frequency, 0., 4. + y_frequency
x_ticks = np.arange(xmin, xmax, x_frequency)
y_ticks = np.arange(ymin, ymax, y_frequency)
ax.set_xticks(x_ticks[x_ticks**2 > 0.005])
ax.set_yticks(y_ticks[y_ticks**2 > 0.005])


# Выводим оба графика
ax.set_title('')
ax.legend()
ax.grid(False)
```

## Задание 7

Постройте хорошие графики нормального распределения
$f = \frac{1}{\sigma \sqrt{2 \pi}} e ^{- \frac{(x - x_{ср})^2}{2 \sigma ^2}}$ для случаев:

1. в одной системе координат постройте два графика $x_{ср} = X = 2$, $\sigma = 1$; $x_{ср} = X = 2$, $\sigma = 0,2$

2. в одной системе координат постройте два графика $x_{ср} = X = 2$, $\sigma = 0,2$; $x_{ср} = X = 4$, $\sigma = 0,2$

```python
fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')
x = np.arange(-1., 5., 0.001)

# Строим первое распределение
X = 2
sigma = 1
y = (np.exp(- (x - X)**2 / (2 * sigma**2))) / (sigma * (2 * np.pi)**0.5)
ax.plot(x, y, linestyle='-', label=rf'$X=2$, $\sigma=1$')

# Строим второе распределение
X = 2
sigma = 0.2
y = (np.exp(- (x - X)**2 / (2 * sigma**2))) / (sigma * (2 * np.pi)**0.5)
ax.plot(x, y, linestyle='--', label=rf'$X=2$, $\sigma=0,2$')

# Рисуем оси
ax.spines[['top', 'right']].set_visible(False)
ax.spines[['top', 'right']].set_visible(False)
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

# Рисуем стрелочки
arrow_fmt = dict(markersize=4, color='black', clip_on=False)
ax.plot((1), (0), marker='>', transform=ax.get_yaxis_transform(), **arrow_fmt)
ax.plot((0), (1), marker='^', transform=ax.get_xaxis_transform(), **arrow_fmt)

# Create 'x' and 'y' labels placed at the end of the axes
ax.set_xlabel('x', size=14, labelpad=0, x=1.)
ax.set_ylabel('F(x)', size=14, labelpad=0, y=1., rotation=0)

# Create custom major ticks to determine position of tick labels
x_frequency, y_frequency = 1., .5 
xmin, xmax, ymin, ymax = -1., 5.+x_frequency, 0., 2. + y_frequency
x_ticks = np.arange(xmin, xmax, x_frequency)
y_ticks = np.arange(ymin, ymax, y_frequency)
ax.set_xticks(x_ticks[x_ticks**2 > 0.005])
ax.set_yticks(y_ticks[y_ticks > 0.1])

# Выводим оба графика
ax.set_title('')
ax.legend()
ax.grid(False);
```
