```python
import numpy as np
import pandas as pd

df = pd.read_excel('kasp.xlsx')
df.head()
```

```python
def get_clear(row):
    return ''

def get_position(row):
    return 'т041559 Программный продукт (ШТ)'

df['Позиция каталога товаров/работ/услуг'] = df.apply(get_position, axis=1)


def get_price(row):
    return row['РРЦ']

df['Цена за единицу'] = df.apply(get_price, axis=1)

df['Цена за единицу со скидкой'] = df.apply(get_clear, axis=1)
df['Количество для скидки'] = df.apply(get_clear, axis=1)

def get_minimal(row):
    try:
        return int(str(row['Band']).split('-')[0])
    except IndexError:
        return int(str(row['Band']))

df['Минимальное количество'] = df.apply(get_minimal, axis=1)


def get_maximal(row):
    try:
        return int(str(row['Band']).split('-')[1])
    except IndexError:
        return int(str(row['Band']))

df['Максимальное количество'] = df.apply(get_maximal, axis=1)


def get_deadline(row):
    return '31.12.2022'

df['Окончание действия оферты'] = df.apply(get_deadline, axis=1)


def get_delivery_time(row):
    return '10 рабочих дней'

df['Сроки поставки'] = df.apply(get_delivery_time, axis=1)


def get_include(row):
    return 'Включены все налоги и пошлины. Не включены услуги по установке и настройке программного обеспечения.'

df['Сведения о включенных (не включённых) в цену расходах'] = df.apply(get_include, axis=1)

def get_name(row):
    name = f'Неисключительное право {row["SaleItemName"][:-11]} [{row["PartNumber"]}]'
    return name

df['Дополнительные сведения'] = df.apply(get_name, axis=1)
    

def get_delivery_type(row):
    return 'Доставка включена в стоимость'

df['Способ доставки'] = df.apply(get_delivery_type, axis=1)

def get_delivery_place(row):
    return 'Муниципальные образования Самарской области'

df['Места доставки'] = df.apply(get_delivery_place, axis=1)
df['Адрес самовывоза'] = df.apply(get_clear, axis=1)
    
df.head()
```

```python
df = df.drop(columns=['PartNumber',
                 'SaleItemName',
                 'Band',
                 'Term',
                 'License type',
                 'РРЦ',
                 'Кол-во',
                 'Рекомендованная стоимость, руб.',
                 'скидка,%',
                 'Ст-ть со скидкой, руб.',
                 'deliv.date',
                 'exp.date',
                 'кол-во дней',
                 'РРЦ за 1 ед., руб./ед.',
                 'Рекомендованная ст-ть дозакупки, руб',
                 'Облагается НДС',
                 'Minsvyaz reg.number',
                 'Minsvyaz reg.date',
                 'Minsvyaz reestr link'])
df.head()
```

```python
df.to_excel('afert.xlsx', index=False)
```
